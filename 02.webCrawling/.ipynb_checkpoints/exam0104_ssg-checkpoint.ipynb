### * 웹 크롤링 SSG

```python
user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.75 Safari/537.36'
```

```python
import requests, bs4

targetUrl = "https://www.ssg.com/search.ssg?target=all&query=%EA%B3%84%EB%9E%80"

resp = requests.get(targetUrl,headers = {'User-Agent':user_agent} )

html = resp.text

bsobj = bs4.BeautifulSoup(html,"html.parser")
bsobj
```

###
- 현재 페이지 80개 제품명 추출

```python
findTags = bsobj.findAll(name="li", attrs={"class":"cunit_t232"})
```

```python
findTags[100]
```

```python
nameTags = bsobj.findAll(name="em", attrs={"class":"tx_ko"})
nameTags
print(len(nameTags))
nameTags[5].text
```

```python
nameList = []
for i in range(0, len(nameTags)): 
    nameList.append(nameTags[i].text)
nameList
```

```python
for i in range(0, len(nameList)):
    print(f"{i} : {nameList[i]}")
```

```python
def isEnglishOrKorean(input_s):
    k_count = 0
    e_count = 0
    for c in input_s:
        if ord('가') <= ord(c) <= ord('힣'):
            k_count+=1
        elif ord('a') <= ord(c.lower()) <= ord('z'):
            e_count+=1
    return 0 if k_count>e_count else 1

#num = isEnglishOrKorean(nameList[11])
#print(nameList[11])
#print(num)

for i in range(0, len(nameList)): 
    num = isEnglishOrKorean(nameList[i])
    #print(f"{i} : {num}")
    if num == 1:
        nameList.remove(nameList[i])
    if nameList[i] == "하림" or nameList[i] == "풀무원":
        nameList.remove(nameList[i])
```

```python
for i in range(0, len(nameList)):
    print(f"{i} : {nameList[i]}")
```

```python
currentItemNameList = []
for i in range(0, 80):
    currentItemNameList.append(nameList[i])
```

```python
for i in range(0, len(currentItemNameList)):
    print(f"{i} : {currentItemNameList[i]}")
```

```python
import pandas as pd
currentItemNameData = pd.DataFrame(currentItemNameList, columns = ["제목이름"])
currentItemNameData
```

```python
# csv 파일로 저장하기
currentItemNameData.to_csv("./currentItemNameData.csv")
```
