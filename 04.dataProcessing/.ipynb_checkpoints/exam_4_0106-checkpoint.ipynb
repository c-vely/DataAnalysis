# 실습
- [불량 데이터 처리]
1. kopo_channel_seasonality_new.csv 자료를 담은 selloutData 변수에서
2. QTY컬럼 음수(반품)인 경우 0, 양수인 경우 기존 QTY 값 유지하는 로직을 적용하여 
3. QTY_NEW 컬럼을 추가하세요

```python
import pandas as pd

selloutData = \
    pd.read_csv("../dataset/kopo_channel_seasonality_new.csv")
```

### A. 데이터 전처리

A-1 타입 통합 및 불량데이터 처리
- 타입 통합

```python
selloutData.dtypes
```

```python
# 리스트에 담고 처리하면 편함
allColumns = list(selloutData.columns)
```

```python
for i in range(0, len(allColumns)):
    selloutData[allColumns[i] ]= selloutData[allColumns[i] ].astype(str)
```

```python
selloutData.QTY = selloutData.QTY.astype(float)
```

```python
selloutData.dtypes
```

### 불량데이터 처리

```python
selloutData.describe()
```

```python
# 함수 
def valuechange(inValue):
    outValue = 0
    
    if inValue < 0:
        outValue = 0
    else:
        outValue = inValue
    return outValue

selloutData["QTY_NEW"] = selloutData.QTY.apply(valuechange)

selloutData
```

```python
# 검증하기
# 1)
selloutData.loc[ selloutData.QTY < 0 ]
```

```python
# 2)
selloutData.loc[ selloutData.QTY > 0 ]
```

```python
# 3)
selloutData.loc[ (selloutData.QTY < 0) &
                  (selloutData.QTY_NEW < 0) ]
```

# 실습
- [데이터 통합]
1. selloutData 자료에서
2. YEAR, WEEK 컬럼 생성하고
3. 53주차를 제거한다

```python
selloutData["YEAR"] = selloutData["YEARWEEK"].str[0:4]
selloutData["WEEK"] = selloutData["YEARWEEK"].str[4:6]
```

```python
exceptWeek = "53"
```

```python
refinedData = selloutData.loc[selloutData.WEEK != exceptWeek ]
```

```python
refinedData
```

# 실습) 대표값 생성
1. refinedData 에서
2. 지역, 상품, 연주차 컬럼순으로 오름차순 정렬하여
3. sortedData 변수에 담으세요

```python
sortKey = ["REGIONID", "PRODUCT", "YEARWEEK"]
```

```python
sortedData = refinedData.sort_values(sortKey)
```

```python
sortedData
```

# 실습)
- [지역, 상품, 연도 별 집계]
1. sortedData 에서 지역, 상품, 연도, 단위 판매량(QTY_NEW) 의 평균 연산 후
2. groupData 변수에 담으세요
3. 이후 컬럼명을 QTY_MEAN로 변경하세요

```python
groupKey = ["REGIONID", "PRODUCT", "YEAR"]
```

```python
groupData = sortedData.groupby(groupKey)["QTY_NEW"].agg(["mean"]).reset_index()
groupData
```

```python
groupData2 = groupData.rename(columns={"mean":"QTY_MEAN"})
```

```python
groupData2
```

# 실습) 추세선 도출 (데이터 병합)
1. refinedSelloutData와 groupData를 [REGIONID, PRODUCT, YEAR] 키로 조인하여
2. mergedData 변수에 아래와 같이 담으세요

```python
joinKey = ["REGIONID", "PRODUCT", "YEAR"]
```

```python
mergedData = pd.merge(left=sortedData,
         right=groupData2,
         how="left",
         on=joinKey)
```

```python
mergedData
```

# 실습)
1. 계절성 지수는 QTY_NEW / QTY_MEAN으로 산출하여
2. SEASONALITY 컬럼을 생성한다

```python
finalData = mergedData.sort_values(by=sortKey)
finalData["SEASONALITY"] = finalData["QTY"] / finalData["QTY_MEAN"]
finalData
```

# 실습)
1. 마지막으로 [지역, 상품, 주차] 별 계절성 지수 평균 값을 조별 산출하여
2. finalResult 변수에 담으세요

```python
groupKey2 = ["REGIONID", "PRODUCT", "WEEK"]
```

```python
seasonData = finalData.groupby(groupKey2)["SEASONALITY"].agg(["mean"]).reset_index()
finalResult = seasonData.rename(columns={"mean":"SEASONALITY"})
finalResult
```
