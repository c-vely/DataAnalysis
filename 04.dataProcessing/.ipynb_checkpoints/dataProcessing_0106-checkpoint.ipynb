### 4. 데이터 조작하기

- productage < 1로 변경하고 그 외에는 기존값을 쓴다

```python
# np.where(조건식, yes값, no값)
import numpy as np

np.where(customerData.PRODUCTAGE < 1, 1, customerData.PRODUCTAGE)
```

```python
# 새 컬럼 만들어서 넣기
customerData["PRODUCTAGE_NEW"] = \
    np.where(customerData.PRODUCTAGE < 1, 1, customerData.PRODUCTAGE)
```

```python
customerData
```

```python
# 검증하기
# 1)
customerData.loc[ (customerData.PRODUCTAGE < 1) &
                  (customerData.PRODUCTAGE_NEW == 1) ] 
```

```python
# 검증하기
# 2)
customerData.loc[ (customerData.PRODUCTAGE < 1) &
                  (customerData.PRODUCTAGE_NEW != 1) ] 
```

```python
# 검증하기
# 3)
customerData.loc[ (customerData.PRODUCTAGE >= 1) &
                  (customerData.PRODUCTAGE_NEW != customerData.PRODUCTAGE) ] 
```

##### 1. np.where 활용 데이터 조작
- PRODUCTAGE 1보다 작은 경우 1
- PRODUCTAGE 1보다 크면서 2보다 작은 경우 2
- PRODUCTAGE 이외는 3

```python
np.where(customerData.PRODUCTAGE < 1, 1, 
        np.where(customerData.PRODUCTAGE < 2, 2, 3))
```

##### 2. loc 활용 데이터 조작
- 데이터프레임명.loc[ (조건식), "대상컬럼" ] = 할당 값

```python
# 단점 : 모든 조건을 다 명시해주야 한다. 
customerData.loc[ customerData.PRODUCTAGE <= 1, "HK_NEW"] = 1
```

```python
# 아래처럼~~
customerData.loc[ customerData.PRODUCTAGE <= 1, "HK_NEW"] = 1
customerData.loc[ (customerData.PRODUCTAGE > 1) &
                  (customerData.PRODUCTAGE <= 2), "HK_NEW"] = 2
customerData.loc[ customerData.PRODUCTAGE > 2 ] = 3
```

```python
customerData
```

##### 3. 함수적용 데이터 조작

```python
inValue = 2.5
outValue = 0
```

```python
if inValue < 1:
    outValue = 1
elif inValue < 2:
    outValue = 2
else:
    outValue = 3 
outValue
```

```python
# 함수 
def valuechange(inValue):
    outValue = 0
    
    if inValue < 1:
        outValue = 1
    elif inValue < 2:
        outValue = 2
    else:
        outValue = 3 
    return outValue
```

```python
customerData["HK_NEW2"] = customerData.PRODUCTAGE.apply(valuechange)
```

```python
customerData
```

### 7. 데이터 그룹별 집계함수

- 예제 1)

```python
columns = ["이름", "성별", "과목", "점수"]
```

```python
testDf = pd.DataFrame( [["깁효과", "남", "수학", 50], 
                        ["깁어진", "남", "수학", 80],
                        ["안민정", "여", "수학", 90],
                        ["안정훈", "여", "국어", 120], 
                        ["김여진", "여", "수학", 95], 
                        ["이동연", "남", "국어", 80]] , columns = columns)
```

```python
groupKey = ["과목"]
```

```python
testDf
```

```python
testDf.groupby(by=groupKey)["점수"].agg(["mean"]).reset_index()
```

```python
groupDf = testDf
```

```python
groupKey = ["성별"]
```

```python
# mean:평균  stdev:표준편차  min:최소값  max:최대값  count:전체 개수
groupDf.groupby(by=groupKey)["점수"].agg(["mean"])
```

```python
groupDf.groupby(by=groupKey)["점수"].agg(["count"]).reset_index()
```

### 9. 데이터 병합

```python
joinMaster = pd.read_csv("../dataset/join_region_master.csv", encoding ="ms949")
```

```python
joinMaster
```

```python
joinSellout = pd.read_csv("../dataset/join_kopo_product_volume.csv")
```

```python
joinSellout
```

```python
joinKey = ["REGIONID"]
```

```python
mergedData = pd.merge(left=joinMaster,     # 기준정보
                      right=joinSellout,   # 실적정보
                      how = "inner",       # innerjoin
                      on = joinKey )
```

```python
mergedData
```

```python
mergedData.loc[:, ["REGIONNAME", "PRODUCT"] ]
```

```python
pd.merge(left=joinSellout,
         right=joinMaster,
         how="left",
         on=joinKey)
```

```python
pd.merge(left=joinSellout,
         right=joinMaster,
         how="right",
         on=joinKey)
```

```python
pd.merge(left=joinSellout,
         right=joinMaster, 
         on= joinKey, 
         how="outer")
```

```python
row1 = ["A99", "ST0010", "201901", 5000]
```

```python
columns = joinSellout.columns
```

```python
chugadata = pd.DataFrame([ row1 ], columns=columns)
```

```python
joinSellout2 = joinSellout.append(chugadata)
```

```python
pd.merge(left=joinSellout2,
         right=joinMaster, 
         on= joinKey, 
         how="outer")
```

```python
# 이름을 달라도 내용이 같으면 조인키로 쓸 수 있음
joinSellout3 = joinSellout2.rename(columns={"REGIONID":"AP2ID"})
```

```python
joinSellout3
```

```python
pd.merge(left=joinSellout3,
         right=joinMaster,
         how="left",
         left_on=["AP2ID"], right_on=["REGIONID"] )
```

```python
# 위, 아래로 합치기
a01Data = joinSellout.loc[joinSellout.REGIONID=="A01"]
a02Data = joinSellout.loc[joinSellout.REGIONID=="A02"]
```

```python
a01Data
```

```python
a02Data
```

```python
# 1) 주의: 컬럼의 이름을 맞춰줘야 한다. 
a01Data.append(a02Data)
```

```python
# 2) 머신러닝에 많이 씀
pd.concat([a01Data, a02Data], axis=0)
```
