### A. 데이터 전처리

A-1 타입 통합 및 불량데이터 처리
- 타입 통합

```python
selloutData.dtypes
```

```python
# 리스트에 담고 처리하면 편함
allColumns = list(selloutData.columns)
```

```python
for i in range(0, len(allColumns)):
    selloutData[allColumns[i] ]= selloutData[allColumns[i] ].astype(str)
```

```python
selloutData.QTY = selloutData.QTY.astype(float)
```

```python
selloutData.dtypes
```

### 불량데이터 처리

```python
selloutData.describe()
```

```python
# 함수 
def valuechange(inValue):
    outValue = 0
    
    if inValue < 0:
        outValue = 0
    else:
        outValue = inValue
    return outValue

selloutData["QTY_NEW"] = selloutData.QTY.apply(valuechange)

selloutData
```

```python
# 검증하기
# 1)
selloutData.loc[ selloutData.QTY < 0 ]
```

```python
# 2)
selloutData.loc[ selloutData.QTY > 0 ]
```

```python
# 3)
selloutData.loc[ (selloutData.QTY < 0) &
                  (selloutData.QTY_NEW < 0) ]
```
